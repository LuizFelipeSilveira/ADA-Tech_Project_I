## 4.   **FUNÇÃO ATRIBUTO_PRODUTO(ARG)**

* **Código bruto:**
```
def atributo_produto(arg): # Coleta os atributos de um produto de acordo com o parâmetro passado
    if arg == 'identificador':
        while True:
            try:
                identificador = str(input('Digite o identificador do produto (5 números inteiros): '))
                if len(identificador) != 5:
                    raise ValueError
                elif identificador.isnumeric() != True:
                    raise TypeError
            except TypeError as e:
                print('Ops! O identificador deve ser um número inteiro de 5 dígitos... Tente novamente.')
            except ValueError as e:
                print('Ops! O identificador deve ser um número inteiro de 5 dígitos... Tente novamente.')
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return identificador

    elif arg == 'nome':
        while True:
            try:
                nome = str(input('Nome do produto: ')).lower().strip()
                if nome == '':
                    raise ValueError
            except ValueError:
                print('Ops! O campo "nome" não pode ser vazio... Tente novamente.')
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return nome

    elif arg == 'marca':
        while True:
            try:
                marca = str(input('Marca do produto: ')).lower().strip()
                if marca == '':
                    raise ValueError
            except ValueError:
                print('Ops! O campo "marca" não pode ser vazio... Tente novamente.')
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return marca

    elif arg == 'preco':
        while True:
            try:
                preco = str(input('Preço do produto (Digite apenas o valor): R$'))
                if ',' in preco:
                    preco = preco.replace(',', '.')
                preco = float(preco)
            except ValueError as e:
                print('Ops! O preço do produto não pode conter letras... Tente novamente.')
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return preco

    elif arg == 'estoque':
        while True:
            try:
                estoque = int(input('Quantidade em estoque (número inteiro): '))
            except TypeError as e:
                print('Ops! O estoque do produto não pode conter letras... Tente novamente.')
            except ValueError as e:
                print('Ops! O estoque do produto deve ser um valor inteiro... Tente novamente.')
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return estoque

    elif arg == 'descricao':
        while True:
            try:
                descricao = str(input(f'Descrição (Aperte ENTER direto para "sem descrição"): ')).lower().strip()
                if descricao == '':
                    descricao = 'sem descrição'
            except:
                print('Ops! Ocorreu um erro inesperado... Tente novamente.')
            else:
                break
        return descricao

    elif arg == 'quantidade':
        while True:
            try:
                quantidade = int(input('Digite quantas unidades desse produto serão vendidas: '))
            except:
                print('Ops! A quantidade precisa ser um número inteiro... Tente novamente.')
            else:
                break
        return quantidade
```


- **Funcionamento:** A função atributo_produto(arg) tem como objetivo coletar dados relacionados ao produto. O parâmetro **arg** deve ser uma **string** com o atributo que deve ser coletado.

Os parâmetros disponíveis são:

4.   "preco"

## 5. **FUNÇÃO VERIFICADOR(IDENTIFICADOR)**
* Código bruto
```
def verificador(identificador):
    lista_indexador = list(map(lambda produto: identificador in produto.keys(), produtos))
    return lista_indexador
```
* Funcionamento:
Basicamente a função faz uma validação de um identificador por meio de lógica booleana da entrada do identificador em relação as chaves dos dicionários na lista "PRODUTOS". Em resumo ela verifica a existência do identificador.

  1. lista_indexador: O lista indexador é uma lista que armazena o valor booleano advindo da comparação do identificador com a chaves do dicionário na lista "produtos". Este processo é feito utilizando uma função lambda que utiliza a variável identificador para comparar com as chaves de um dicionario chamado "produto". Caso identificador = chave, temos como saída "True", caso o contrário, a saída será False.
  ```
  lambda produto: identificador in produto.keys()
  ```
  2. Para que seja feito em isto em todos os dicionários de "produtos", utilizamos a função "map", o que nos dá a condição d percorrer todos os dicionários.

def ler_arquivo(nome_arquivo, tabela):
    tabela.clear()
    with open(nome_arquivo, 'r') as arquivo:
        planilha = csv.reader(arquivo, delimiter = ';', lineterminator='\n')
        for linha in planilha:
            tabela.append(linha)
        print(f'Arquivo {arquivo} encontrado e lido com sucesso.')

```

- **Funcionamento:** Essa função faz a leitura do arquivo, passando como o nosso sistema deve ler o arquivo, delimitando ele com ';' e terminado a linha pulando pra outra linha. Criamos um loop for para pegar cada linha existente na planilha e adicionamos na tabela.

- Colocamos uma mensagem amigavel no print falando que o arquivo foi encontrado e lido sucesso.


## 15. EXPORTAR ARQUIVO

* **Código bruto:**

```
